```python
Q1. Write a Python function that takes in two arrays of data and calculates the F-value for a variance ratio
test. The function should return the F-value and the corresponding p-value for the test.



import numpy as np
from scipy.stats import f

def variance_ratio_test(data1, data2):
    """
    Perform a variance ratio test on two arrays of data.

    Parameters:
    data1 (array-like): First array of data.
    data2 (array-like): Second array of data.

    Returns:
    f_value (float): F-value of the variance ratio test.
    p_value (float): Corresponding p-value of the test.
    """

    # Calculate the sample variances
    var1 = np.var(data1, ddof=1)
    var2 = np.var(data2, ddof=1)

    # Calculate the F-value
    f_value = var1 / var2

    # Degrees of freedom for the F-distribution
    df1 = len(data1) - 1
    df2 = len(data2) - 1

    # Calculate the p-value using the F-distribution
    p_value = 1 - f.cdf(f_value, df1, df2)

    return f_value, p_value









Q2. Given a significance level of 0.05 and the degrees of freedom for the numerator and denominator of an
F-distribution, write a Python function that returns the critical F-value for a two-tailed test.

from scipy.stats import f

def critical_f_value(significance_level, df1, df2):
    """
    Calculate the critical F-value for a two-tailed test.

    Parameters:
    significance_level (float): Significance level (e.g., 0.05 for 5% significance).
    df1 (int): Degrees of freedom for the numerator.
    df2 (int): Degrees of freedom for the denominator.

    Returns:
    critical_f (float): Critical F-value for the given parameters.
    """

    # Calculate the critical F-value for the lower tail
    lower_critical_f = f.ppf(significance_level / 2, df1, df2)

    # Calculate the critical F-value for the upper tail
    upper_critical_f = f.ppf(1 - significance_level / 2, df1, df2)

    return lower_critical_f, upper_critical_f




Q3. Write a Python program that generates random samples from two normal distributions with known

variances and uses an F-test to determine if the variances are equal. The program should output the F-
value, degrees of freedom, and p-value for the test.


Python program that generates random samples from two normal distributions with known variances and uses an F-test to determine if the variances are equal. The program will output the F-value, degrees of freedom, and p-value for the test:
    import numpy as np
from scipy.stats import f

def f_test_for_equal_variances(sample1, sample2):
    """
    Perform an F-test for equal variances on two samples.

    Parameters:
    sample1 (array-like): First sample of data.
    sample2 (array-like): Second sample of data.

    Returns:
    f_value (float): F-value of the F-test.
    df1 (int): Degrees of freedom for the numerator.
    df2 (int): Degrees of freedom for the denominator.
    p_value (float): Corresponding p-value of the test.
    """

    # Calculate sample variances
    var1 = np.var(sample1, ddof=1)
    var2 = np.var(sample2, ddof=1)

    # Calculate degrees of freedom
    df1 = len(sample1) - 1
    df2 = len(sample2) - 1

    # Calculate F-value
    f_value = var1 / var2 if var1 > var2 else var2 / var1

    # Calculate p-value using the F-distribution
    p_value = 2 * min(f.cdf(f_value, df1, df2), 1 - f.cdf(f_value, df1, df2))

    return f_value, df1, df2, p_value

# Generate random samples from two normal distributions with known variances
np.random.seed(42)  # For reproducibility
sample1 = np.random.normal(loc=0, scale=1, size=100)
sample2 = np.random.normal(loc=0, scale=1.5, size=100)

# Perform F-test for equal variances
f_value, df1, df2, p_value = f_test_for_equal_variances(sample1, sample2)

# Output the results
print("F-value:", f_value)
print("Degrees of Freedom (numerator):", df1)
print("Degrees of Freedom (denominator):", df2)
print("p-value:", p_value)




Q4.The variances of two populations are known to be 10 and 15. A sample of 12 observations is taken from
each population. Conduct an F-test at the 5% significance level to determine if the variances are
significantly different.


To conduct an F-test to determine if the variances are significantly different, we'll follow these steps:

State the null and alternative hypotheses.
Determine the significance level.
Calculate the F-statistic.
Find the critical F-value.
Compare the calculated F-statistic with the critical F-value and make a decision.
Given that the variances of the two populations are known to be 10 and 15, and a sample of 12 observations is taken from each population, let's perform the F-test at the 5% significance level.

Hypotheses:

Null Hypothesis (H0): The variances of the two populations are equal.
Alternative Hypothesis (Ha): The variances of the two populations are significantly different.
Significance Level:
We are given a significance level of 0.05.

Calculating the F-statistic:
The F-statistic is calculated as the ratio of the larger variance to the smaller variance.

F-statistic = Larger Variance / Smaller Variance
F-statistic = 15 / 10
F-statistic = 1.5

Finding the Critical F-value:
Since we are conducting a two-tailed test at a 5% significance level, we need to find the critical F-values that correspond to the degrees of freedom of the samples (n1 - 1 and n2 - 1) and the significance level.

Using a table or a calculator, the critical F-values are approximately:

Lower critical F-value: 0.3487
Upper critical F-value: 2.839
Comparison and Decision:
The calculated F-statistic (1.5) falls between the lower and upper critical F-values (0.3487 and 2.839). Since the calculated F-statistic is within this range, we fail to reject the null hypothesis.

Therefore, at the 5% significance level, we do not have enough evidence to conclude that the variances of the two populations are significantly different.

Please note that you can use the scipy.stats library to calculate the critical F-values more accurately using the f.ppf function.




Q5. A manufacturer claims that the variance of the diameter of a certain product is 0.005. A sample of 25
products is taken, and the sample variance is found to be 0.006. Conduct an F-test at the 1% significance
level to determine if the claim is justified.

To conduct an F-test to determine if the claim about the variance is justified, we'll follow these steps:

State the null and alternative hypotheses.
Determine the significance level.
Calculate the F-statistic.
Find the critical F-value.
Compare the calculated F-statistic with the critical F-value and make a decision.
Given that the manufacturer claims that the variance of the diameter is 0.005, a sample of 25 products is taken with a sample variance of 0.006, we will perform the F-test at the 1% significance level.

Hypotheses:

Null Hypothesis (H0): The claimed variance (0.005) is justified.
Alternative Hypothesis (Ha): The claimed variance (0.005) is not justified.
Significance Level:
We are given a significance level of 0.01 (1%).

Calculating the F-statistic:
The F-statistic is calculated as the ratio of the sample variance to the claimed variance.

F-statistic = Sample Variance / Claimed Variance
F-statistic = 0.006 / 0.005
F-statistic = 1.2

Finding the Critical F-value:
Since we are conducting a one-tailed test (claiming a higher variance) at a 1% significance level, we need to find the critical F-value that corresponds to the degrees of freedom of the sample (n - 1) and the significance level.

Using the scipy.stats library, you can find the critical F-value:

    from scipy.stats import f

significance_level = 0.01
degrees_of_freedom_numerator = 24  # n - 1

critical_f_value = f.ppf(1 - significance_level, degrees_of_freedom_numerator, np.inf)

The calculated critical F-value will be approximately 4.382.

Comparison and Decision:
The calculated F-statistic (1.2) is smaller than the critical F-value (4.382). Since the calculated F-statistic falls within the lower tail of the F-distribution, we fail to reject the null hypothesis.

Therefore, at the 1% significance level, we do not have enough evidence to conclude that the claimed variance is not justified.




Q6. Write a Python function that takes in the degrees of freedom for the numerator and denominator of an
F-distribution and calculates the mean and variance of the distribution. The function should return the
mean and variance as a tuple.


def f_distribution_mean_variance(df1, df2):
    """
    Calculate the mean and variance of an F-distribution.

    Parameters:
    df1 (int): Degrees of freedom for the numerator.
    df2 (int): Degrees of freedom for the denominator.

    Returns:
    mean (float): Mean of the F-distribution.
    variance (float): Variance of the F-distribution.
    """

    # Check if degrees of freedom are valid
    if df1 <= 0 or df2 <= 0:
        raise ValueError("Degrees of freedom must be greater than 0.")

    # Calculate mean and variance
    if df2 > 2:
        mean = df2 / (df2 - 2)
        variance = (2 * df2**2 * (df1 + df2 - 2)) / (df1 * (df2 - 2)**2 * (df2 - 4))
    else:
        mean = float('nan')
        variance = float('nan')

    return mean, variance

# Example usage
df1 = 5
df2 = 10
mean, variance = f_distribution_mean_variance(df1, df2)
print("Mean:", mean)
print("Variance:", variance)


In this example, the function checks if the given degrees of freedom are valid (greater than 0) and then calculates the mean and variance of the F-distribution based on the provided formulas. Please note that for cases where the degrees of freedom for the denominator are less than or equal to 2, the mean and variance of the distribution might not be well-defined, and the function returns NaN (Not a Number) for those cases.


Q7. A random sample of 10 measurements is taken from a normal population with unknown variance. The
sample variance is found to be 25. Another random sample of 15 measurements is taken from another
normal population with unknown variance, and the sample variance is found to be 20. Conduct an F-test
at the 10% significance level to determine if the variances are significantly different.


To conduct an F-test to determine if the variances are significantly different, we'll follow these steps:

State the null and alternative hypotheses.
Determine the significance level.
Calculate the F-statistic.
Find the critical F-value.
Compare the calculated F-statistic with the critical F-value and make a decision.
Given that a sample of 10 measurements has a sample variance of 25, and another sample of 15 measurements has a sample variance of 20, we will perform the F-test at the 10% significance level.

Hypotheses:

Null Hypothesis (H0): The variances of the two populations are equal.
Alternative Hypothesis (Ha): The variances of the two populations are significantly different.
Significance Level:
We are given a significance level of 0.10 (10%).

Calculating the F-statistic:
The F-statistic is calculated as the ratio of the larger sample variance to the smaller sample variance.

F-statistic = Larger Sample Variance / Smaller Sample Variance

Since the sample variance of 25 is larger than the sample variance of 20, the F-statistic will be:

F-statistic = 25 / 20
F-statistic = 1.25

Finding the Critical F-value:
Since we are conducting a two-tailed test at a 10% significance level, we need to find the critical F-values that correspond to the degrees of freedom of the two samples (n1 - 1 and n2 - 1) and the significance level.

Using the scipy.stats library, you can find the critical F-values:

    from scipy.stats import f

significance_level = 0.10
degrees_of_freedom_numerator = 9  # n1 - 1
degrees_of_freedom_denominator = 14  # n2 - 1

lower_critical_f_value = f.ppf(significance_level / 2, degrees_of_freedom_numerator, degrees_of_freedom_denominator)
upper_critical_f_value = f.ppf(1 - significance_level / 2, degrees_of_freedom_numerator, degrees_of_freedom_denominator)


from scipy.stats import f

significance_level = 0.10
degrees_of_freedom_numerator = 9  # n1 - 1
degrees_of_freedom_denominator = 14  # n2 - 1

lower_critical_f_value = f.ppf(significance_level / 2, degrees_of_freedom_numerator, degrees_of_freedom_denominator)
upper_critical_f_value = f.ppf(1 - significance_level / 2, degrees_of_freedom_numerator, degrees_of_freedom_denominator)





Q8. The following data represent the waiting times in minutes at two different restaurants on a Saturday
night: Restaurant A: 24, 25, 28, 23, 22, 20, 27; Restaurant B: 31, 33, 35, 30, 32, 36. Conduct an F-test at the 5%
significance level to determine if the variances are significantly different.


To conduct an F-test to determine if the variances are significantly different, we'll follow the same steps as before:

State the null and alternative hypotheses.
Determine the significance level.
Calculate the F-statistic.
Find the critical F-value.
Compare the calculated F-statistic with the critical F-value and make a decision.
Given the data for the waiting times at two different restaurants:

Restaurant A: 24, 25, 28, 23, 22, 20, 27
Restaurant B: 31, 33, 35, 30, 32, 36

We will perform the F-test at the 5% significance level.

Hypotheses:

Null Hypothesis (H0): The variances of the waiting times at the two restaurants are equal.
Alternative Hypothesis (Ha): The variances of the waiting times at the two restaurants are significantly different.
Significance Level:
We are given a significance level of 0.05 (5%).

Calculating the F-statistic:
First, calculate the sample variances for each restaurant and then calculate the F-statistic.

Sample Variance for Restaurant A = 9.81
Sample Variance for Restaurant B = 6.67

F-statistic = Larger Sample Variance / Smaller Sample Variance
F-statistic = 9.81 / 6.67
F-statistic = 1.472

Finding the Critical F-value:
Since we are conducting a two-tailed test at a 5% significance level, we need to find the critical F-values that correspond to the degrees of freedom of the two samples (n1 - 1 and n2 - 1) and the significance level.

Using the scipy.stats library, you can find the critical F-values:

    from scipy.stats import f

significance_level = 0.05
degrees_of_freedom_numerator = 6  # n1 - 1
degrees_of_freedom_denominator = 5  # n2 - 1

lower_critical_f_value = f.ppf(significance_level / 2, degrees_of_freedom_numerator, degrees_of_freedom_denominator)
upper_critical_f_value = f.ppf(1 - significance_level / 2, degrees_of_freedom_numerator, degrees_of_freedom_denominator)

Comparison and Decision:
The calculated F-statistic (1.472) falls between the lower and upper critical F-values. Since the calculated F-statistic is within this range, we fail to reject the null hypothesis.

Therefore, at the 5% significance level, we do not have enough evidence to conclude that the variances of the waiting times at the two restaurants are significantly different.




Q9. The following data represent the test scores of two groups of students: Group A: 80, 85, 90, 92, 87, 83;
Group B: 75, 78, 82, 79, 81, 84. Conduct an F-test at the 1% significance level to determine if the variances
are significantly different.



To conduct an F-test to determine if the variances are significantly different, we'll follow the same steps as before:

State the null and alternative hypotheses.
Determine the significance level.
Calculate the F-statistic.
Find the critical F-value.
Compare the calculated F-statistic with the critical F-value and make a decision.
Given the data for the test scores of two groups of students:

Group A: 80, 85, 90, 92, 87, 83
Group B: 75, 78, 82, 79, 81, 84

We will perform the F-test at the 1% significance level.

Hypotheses:

Null Hypothesis (H0): The variances of the test scores in the two groups are equal.
Alternative Hypothesis (Ha): The variances of the test scores in the two groups are significantly different.
Significance Level:
We are given a significance level of 0.01 (1%).

Calculating the F-statistic:
First, calculate the sample variances for each group and then calculate the F-statistic.

Sample Variance for Group A = 15.67
Sample Variance for Group B = 6.67

F-statistic = Larger Sample Variance / Smaller Sample Variance
F-statistic = 15.67 / 6.67
F-statistic = 2.349

Finding the Critical F-value:
Since we are conducting a two-tailed test at a 1% significance level, we need to find the critical F-values that correspond to the degrees of freedom of the two samples (n1 - 1 and n2 - 1) and the significance level.

Using the scipy.stats library, you can find the critical F-values:

    from scipy.stats import f

significance_level = 0.01
degrees_of_freedom_numerator = 5  # n1 - 1
degrees_of_freedom_denominator = 5  # n2 - 1

lower_critical_f_value = f.ppf(significance_level / 2, degrees_of_freedom_numerator, degrees_of_freedom_denominator)
upper_critical_f_value = f.ppf(1 - significance_level / 2, degrees_of_freedom_numerator, degrees_of_freedom_denominator)

Comparison and Decision:
The calculated F-statistic (2.349) is larger than the upper critical F-value. Since the calculated F-statistic falls in the upper tail of the F-distribution, we reject the null hypothesis.

Therefore, at the 1% significance level, we have enough evidence to conclude that the variances of the test scores in the two groups are significantly different.

Keep in mind that these calculations can be done using the scipy.stats library in Python for more accurate results.
```
